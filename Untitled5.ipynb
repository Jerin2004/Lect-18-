# **Introduction to Matplotlib and Seaborn**

```python
import matplotlib.pyplot as plt
 # Data
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 # Plot
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1)
plt.plot(y2)
plt.show
```

```python
# Data
x= [1,2,3,4,5]
y= [2,3,5,7,11]
# Creating Scatter Plot
plt.scatter(x, y)
plt.show()
```

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 1, 10]
# Creating Bar Plot
plt.bar(categories, values)
plt.title('Simplr Bar Plot')
plt.show()
```

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 1, 10]
# Creating Bar Plot
plt.barh(categories, values)
plt.title('Simplr Bar Plot')
plt.show()
```

```python
# Data
data = [1,2,2,3,3,3,4,4,4,4]

# Creating Histogram
plt.hist(data)
plt.show()
```

# **Adding Annotation with Matplotlib**

```python
# Data
x = [1,2,3,4,5]
y = [2,3,5,7,11]

# Creating a plot with annotation
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('plot with annotation')
plt.annotate('Peak', xy=(5,11), xytext=(5,15), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

# Creating a Simple Line Plot with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
x = [1,2,3,4,5]
y = [2,3,5,7,11]

# Creating a simple line plot
sns.lineplot(x=x, y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```
